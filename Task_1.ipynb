# 4.Explain the difference between stemming and lemmatization.

Stemming reduces a word to its root form by removing parts of the word using simple rules.

Examples:

playing → play

studies → studi

running → run


Sometimes stemming produces words that are not actual dictionary words.



**Lemmatization** converts words into their meaningful base form using vocabulary and grammar rules.

Examples:

playing → play

studies → study

better → good


Lemmatization usually produces valid dictionary words and is more accurate than stemming.

## **Task 2:Build Advanced Preprocessing Function**

```python
import re
from typing import List
```

```python
KEEP_WORDS = {"no", "not"}
```

```python
def remove_repeated_chars(word):
    """
    soooooo -> so
    goooood -> good
    """
    return re.sub(r'(.)\1{2,}', r'\1', word)
```

```python
def preprocess_text(text):

    # Error handling
    if not isinstance(text, str):
        return [], ""

    if not text.strip():
        return [], ""

    # Lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', ' ', text)

    # Remove email patterns
    text = re.sub(r'\S+@\S+', ' ', text)

    # Remove numbers
    text = re.sub(r'\d+', ' ', text)

    # Remove emojis and special chars
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)

    # Remove repeated characters
    text = " ".join(
        remove_repeated_chars(word)
        for word in text.split()
    )

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = []

    for word in text.split():

        if word in KEEP_WORDS:
            tokens.append(word)

        elif len(word) > 2:
            tokens.append(word)

    clean_sentence = " ".join(tokens)

    return tokens, clean_sentence
```

# **Task 3: Stress Testing**

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
results = []

for sentence in test_sentences:

    tokens, clean_sentence = preprocess_text(sentence)

    results.append({
        "original": sentence,
        "tokens": tokens,
        "clean_sentence": clean_sentence
    })

    print("="*60)
    print("Original :", sentence)
    print("Tokens   :", tokens)
    print("Cleaned  :", clean_sentence)
```

# **Task 4: Token Analytics**

```python
def token_analytics(tokens):

    total_tokens = len(tokens)

    unique_tokens = len(set(tokens))

    avg_length = (
        sum(len(word) for word in tokens) / total_tokens
        if total_tokens else 0
    )

    return {
        "total_tokens": total_tokens,
        "unique_tokens": unique_tokens,
        "average_token_length": round(avg_length, 2)
    }
```

```python
for item in results:

    analytics = token_analytics(item["tokens"])

    print("\nSentence:")
    print(item["clean_sentence"])

    print(analytics)
```

# **Task 6: Full Pipeline**

```python
def full_pipeline(text_list):

    all_tokens = []
    clean_sentences = []

    for text in text_list:

        tokens, clean_sentence = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(clean_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
output = full_pipeline(test_sentences)

print(output)
```
